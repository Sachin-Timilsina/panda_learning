```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Depression Student Dataset.csv')
df.head()
```

## Types of Columns:
- Numerical Columns:
    - Age
    - Academic Pressure
    - Study Satisfaction
    - Study Hours
    - Financial Stress
- Categorical Columns:
    -  Gender
    -  Sleep Duration
    -  Dietary Habits
    -  Suicidal Thoughts
    -  Family History
    -  Depression

```python
df.rename(columns={'Have you ever had suicidal thoughts ?':'Suicidal Thoughts'}, inplace=True)
df.rename(columns={'Family History of Mental Illness':'Family History'}, inplace=True)
df.head()
```

```python
# Shape of the dataset
print("Shape of the dataset:", df.shape)
```

```python
categorical_columns = ['Gender', 'Sleep Duration', 'Dietary Habits', 'Suicidal Thoughts', 'Family History', 'Depression']
numerical_columns = ['Age', 'Academic Pressure', 'Study Satisfaction', 'Study Hours', 'Financial Stress']

# Unique Values in Categorical Columns
for categorical_column in categorical_columns:
    print(f"{categorical_column} = {df[categorical_column].unique()}")  # Print unique values of the column
    print()
```

```python
# Value Range in Numerical Columns
for numerical_column in numerical_columns:
    print(f"{numerical_column} : Max Value = {df[numerical_column].max()} and Min Value = {df[numerical_column].min()}")
    print()
```

```python
# Extract min and max values for each numerical column
min_values = [df[col].min() for col in numerical_columns]
max_values = [df[col].max() for col in numerical_columns]

# Create the plot
fig, ax = plt.subplots(figsize=(12, 6))
index = range(len(numerical_columns))

bar_width = 0.35
plt.bar(index, min_values, bar_width, label='Min')
plt.bar([i + bar_width for i in index], max_values, bar_width, label='Max')

plt.xlabel('Numerical Columns')
plt.ylabel('Values')
plt.title('Min and Max Values of Numerical Columns')
plt.xticks([i + bar_width / 2 for i in index], numerical_columns)
plt.legend()

plt.show()
```

```python
# Checking for missing values
df.isnull().sum()
```

## Bivariate Analysis

### Sleep Duration - Depression Analysis
***Conclusions***
- From the analysis it becomes clear that those who slept 8 or more than 8 hours were less likely to have depression.
- According to the data, those who have less than 8 hours sleep were more likely to have depression.

```python
contingency_table = pd.crosstab(df['Depression'], df['Sleep Duration'], normalize='columns')*100
print(contingency_table)
```

```python
# Create a heatmap
cross_tab = pd.crosstab(df['Sleep Duration'], df['Depression'])
sns.heatmap(cross_tab, annot=True, cmap='Blues', fmt='d')
plt.xlabel('Depression')
plt.ylabel('Sleep Duration')
plt.title('Heatmap of Sleep Duration vs Depression')
plt.show()
```

```python
# Bar plot
sns.countplot(data=df, x='Sleep Duration', hue='Depression')
plt.title("Sleep Duration vs Depression")
plt.xlabel("Sleep Duration")
plt.ylabel("Count")
plt.show()
```

### Dietary Habits - Depression Analysis:
***Conclusions***
- According to the analysis, those who have healthy dietary habits were less likely to have depression
- Those who have unhealthy dietary habits were more likely to have depression.
- For moderate dietary habits there's half/half chance of depression and no depression.

```python
contingency_table = pd.crosstab(df['Depression'], df['Dietary Habits'], normalize='columns')*100
print(contingency_table)
```

```python
# Create a heatmap
cross_tab = pd.crosstab(df['Dietary Habits'], df['Depression'])
sns.heatmap(cross_tab, annot=True, cmap='Blues', fmt='d')
plt.xlabel('Depression')
plt.ylabel('Dietary Habits')
plt.title('Heatmap of Dietary Habits vs Depression')
plt.show()
```

```python
# Bar plot
sns.countplot(data=df, x='Dietary Habits', hue='Depression')
plt.title("Dietary Habits vs Depression")
plt.xlabel("Dietary Habits")
plt.ylabel("Count")
plt.show()
```

### Study Satisfaction - Suicidal Thoughts Analysis:
***Conclusions***
- From the visualization, it shows that higher study satisfaction leads to less suicidal thoughts.

```python
# Descriptive Statistics
group_stats = df.groupby('Suicidal Thoughts')['Study Satisfaction'].describe()
print(group_stats)
```

```python
df[df['Suicidal Thoughts'] == 'Yes']['Study Satisfaction'].plot(kind='kde', label='Having Suicidal Thoughts')
df[df['Suicidal Thoughts'] == 'No']['Study Satisfaction'].plot(kind='kde', label='Having no Suicidal Thoughts')

plt.legend()
plt.show()
```
